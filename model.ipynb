```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("Online Retial 2.csv")
```

```python
df
```

```python
df.isna().sum()
```

```python
# Dropping 3k Recency and LastOrderDate null values
df.dropna(inplace=True)
```

```python
df.columns
```

```python
# Get the total spendings for each customer

# Creating total price per each order
df['TotalPrice'] = df['UnitPrice'] * df['Quantity']

# Creating dataframe containing total spending per CustomerID
TotalCustomerSpending = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
TotalCustomerSpending.columns = ['CustomerID', 'TotalCustomerSpending']

# Merging two dataframes
df = pd.merge(df, TotalCustomerSpending, how='left')
df.sample(10)
```

```python
# Depricting total spendings per customer
TotalCustomerSpending = df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
TotalCustomerSpending = TotalCustomerSpending.sort_values('TotalPrice', ascending=False)

# Getting the top customer for a visible graph
px.bar(TotalCustomerSpending, x='CustomerID', y='TotalPrice')
```
